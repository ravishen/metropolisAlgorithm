### IMPORT FILES

```python
import random
import math
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import pylab
```

### RANDOM GENERATOR FROM [0,1]

```python
def randomGenerator():
    num = random.randint(0,100)
    num/=100
    return num
```

### CREATING 2D MATRIX CALLED BOX SURROUNDED BY 0

```python
print("Enter value of L")
l = int(input())
box = []

for i in range (l+2):
    box.append([])
    for j in range (l+2):
        if(i==l+1 or j==0 or i==0 or j==l+1 ):
            box[i].append(0)
            continue
        if(randomGenerator() >= 0.5):
            box[i].append(1)
        else:
            box[i].append(-1)

print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in box]))
```

### ARRAY AS IMAGE

```python
def image_box():
    myarray = np.asarray(box)
    myarray
    im=myarray.reshape(l+2,l+2)
    plt.imshow(im)
    pylab.savefig('foo.png')
image_box()
```

### ENERGY CALCULATIOIN OF ISING MODEL

```python
s = [[0 for x in range(l+2)] for y in range(l+2)]
energy = 0
def total_energy():
    global energy
    for i in range (1,l+1):
        for j in range (1,l+1):
            s[i][j]=(box[i][j])*(box[i+1][j]+box[i-1][j]+box[i][j-1]+box[i][j+1]) 
            energy+=s[i][j]
    print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in s]))
    energy/=2
    return energy


print(total_energy())
print(energy)
```

### MAGNETIZATION CALCULATION

```python
magnetization=0
def total_magnetization():
    global magnetization
    sum=0
    for i in range(l+2):
        for j in range(l+2):
            sum+=s[i][j]
        magnetization=sum/(l*l)
    return magnetization

print(total_magnetization())
print(magnetization)
```

### CALCULATION OF DIFFERENCE IN ENERGY

```python
def diff_energy(i,j):
    difference = -2*(box[i][j])*(box[i+1][j]+box[i-1][j]+box[i][j-1]+box[i][j+1])
    return difference
```

### CHOOSING A RANDON SITE AND DIFFERENCE IN ENERGY

```python
pos_x=0
pos_y=0
def rand_site():
    global pos_x
    global pos_y
    pos_x = 1+int(randomGenerator()*(l-1))
    pos_y = 1+int(randomGenerator()*(l-1))
    print(pos_x,pos_y)
    print (diff_energy(pos_x,pos_y))
    return (diff_energy(pos_x,pos_y))
rand_site()
```

### SETTING VALUE OF TEMPRATURE

```python
temp = int(input())
k=1.380*math.pow(10,-23)
b = 1/(k*temp)
print(b)
```

### SHOULD INVERT THE SPIN OR NOT 

## [METROPOLIS ALGORITHM ]

```python
def metropolis():
    global energy
    global magnetization
    global pos_x
    global pos_y
    delta = diff_energy(pos_x,pos_y)
    print("delta = " + str(delta))
    if (delta < 0):
        print("spin at " +str(pos_x),str(pos_y)+"is inverted as difference in energy in negative")
        box[pos_x][pos_y]=-box[pos_x][pos_y]
        energy+=delta
        print("new energy ="+ str(energy))
        magnetization+=2*box[pos_x][pos_y]
        print("new magnetization ="+ str(magnetization))
    else:
        a = randomGenerator()
        if(a<math.exp(-b*diff_energy(pos_x,pos_y))):
                print("spin at " +str(pos_x),str(pos_y)+"is inverted as "+ str(a) +"<"+ str(math.exp(-b*diff_energy(pos_x,pos_y))))
                box[pos_x][pos_y]=-box[pos_x][pos_y]
                energy+=delta
                print("new energy ="+ str(energy))
                magnetization+=2*box[pos_x][pos_y]
                print("new magnetization ="+ str(magnetization))
        else:
            print("spin should not be inverted")

metropolis()
print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in box]))
print("\n\n")
print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in s]))
```

### SAVING IMAGE

```python
def save_state_as_image(i):
    myarray = np.asarray(box)
    myarray
    im=myarray.reshape(l+2,l+2)
    plt.imshow(im)
    pylab.savefig('images_iteration/'+str(i)+'.png')
save_state_as_image(1)
```

### NO OF ITERATION 

```python
no_of_iteration = int(input())
energy_plot =[]
iterator = [i for i in range(no_of_iteration)]
for i in range(no_of_iteration):
    rand_site()
    metropolis()
    if(i%10==0):
        save_state_as_image(i)
    energy_plot.append(energy)
numpy_energy_array = np.array(energy)
plt.grid(True)
%matplotlib notebook
plt.plot(iterator,energy_plot)
pylab.savefig('graph.png')
```
